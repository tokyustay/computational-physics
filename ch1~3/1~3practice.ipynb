```python
import numpy as np
import matplotlib.pyplot as plt

v0_kmh = 100
v0 = v0_kmh * 1000 / 3600
g = 10
theta_degrees = 45
theta = np.radians(theta_degrees)

t = np.linspace(0, 2*v0*np.sin(theta)/g, num=500)

x = v0 * np.cos(theta) * t
y = (v0 * np.sin(theta) * t) - (0.5 * g * t**2)


plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Projectile Motion Trajectory')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import io

seoul_temp1_data = pd.read_csv(io.StringIO(uploaded['seoul_temp1 (2).txt'].decode('utf-8')), sep=" ", names=["Day", "T1", "T2", "T3"], header=None)

seoul_temp_data = pd.read_csv(io.StringIO(uploaded['seoul_temp (3).txt'].decode('utf-8')), sep=" ", names=["Day", "T1", "T2", "T3"], header=None)

seoul_temp1_data.head()

seoul_temp1_data['T_Ave'] = seoul_temp1_data[['T1', 'T2', 'T3']].sum(axis=1) / 3

seoul_temp1_ave = seoul_temp1_data[['Day', 'T_Ave']]

seoul_temp1_ave['Day'] = pd.to_datetime(seoul_temp1_ave['Day'])

yearly1_ave_temp = seoul_temp1_ave.groupby(seoul_temp1_ave['Day'].dt.year)['T_Ave'].mean().reset_index()

#####################

seoul_temp_data['T_Ave'] = seoul_temp_data[['T1', 'T2', 'T3']].sum(axis=1) / 3

seoul_temp_ave = seoul_temp_data[['Day', 'T_Ave']]

seoul_temp_ave['Day'] = pd.to_datetime(seoul_temp_ave['Day'])

yearly_ave_temp = seoul_temp_ave.groupby(seoul_temp_ave['Day'].dt.year)['T_Ave'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

total_yearly_ave = pd.concat([yearly1_ave_temp, yearly_ave_temp])

plt.figure(figsize=(15, 7))
plt.plot(total_yearly_ave['Day'], total_yearly_ave['T_Ave'], label='Yearly Average Temperature', marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Yearly Average Temperature in Seoul')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
