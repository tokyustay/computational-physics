```python
# This Python code generates a scatter plot with varying marker sizes and colors based on randomly generated data.
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(0, 1, 10)
# Generate 10 random data points for x coordinates from a normal distribution centered at 0 with a standard deviation of 1
y = np.random.normal(0, 1, 10)
# Generate 10 random data points for y coordinates in a similar manner as x
sizes = np.random.uniform(10, 100, 10)
# Generate 10 random sizes for the scatter plot markers, ranging uniformly between 10 and 100
colors = np.random.uniform(0, 1, 10)
# Generate 10 random colors for the scatter plot markers, with values ranging uniformly between 0 and 1
# Create a scatter plot with the generated x and y data points
plt.scatter(x, y, s=sizes, c=colors)
# 's' parameter specifies the sizes of the markers
# 'c' parameter specifies the colors of the markers
plt.xlabel('x')
plt.ylabel('y')
# Add a color bar to the plot, which indicates the color scale used for the markers
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the wave parameters
wavelength = 15.0 # Set the wave's wavelength to 15 units. Here unit is cm.
k = np.pi * 2 / wavelength # Calculate the wave number based on the wavelength.
# Define the grid parameters
side = 100 # Assume one side of the grid is 100 units long.
points = 500 # Divide each side of the grid into 500 points for a detailed calculation.
spacing = side / points # Calculate the spacing between points given the side and number of points.
# Initialize an empty array to store the wave data
A = np.empty([points, points], float) # Create an empty matrix for recording data.
# Generate the wave pattern
for i in range(points):
 y = spacing * i # Calculate the y-coordinate for each row.
 for j in range(points):
   x = spacing * j  # Calculate the x-coordinate for each column.
   r = np.sqrt(x**2 + y**2) # Calculate the distance from the origin (0,0) to the point (x,y).
   A[i, j] = np.sin(k * r) # Calculate the wave's amplitude at each point using a circular wave formula.
# Plot the wave pattern
plt.imshow(A, origin="lower", extent=[0, side, 0, side]) # Display the wave pattern with the origin at the bottom left.
plt.xlabel("x (cm)") # Label the x-axis in centimeters.
plt.ylabel("y (cm)") # Label the y-axis in centimeters.
plt.colorbar() # Show a color bar indicating the amplitude of the wave.
plt.gray() # Display the image in grayscale. Without this, the default colormap is 'viridis,' designed to be perceptually uniform.
#plt.jet()
#plt.hsv()
#plt.viridis()
#plt.hot()
#plt.spectral()
#plt.bone()
plt.show() # Display the plot.
```

```python
# Using if to print at every 100th iteration
N = 10000
sum = 0.0
for n in range(N):
  x = 4 * (-1)**n / (2 * n + 1)
  sum += x # Add each term to the sum
 # Check if the current iteration is a multiple of 100
  if (n % 100 == 0):
 # If true, print the current iteration number and the sum
     print(n, sum)
```

```python
import numpy as np
import matplotlib.pyplot as plt
eps = 1e-3
sum = 0.0
n = 0
n_list = [] # initialize list for x axis containing the iteration number
y_list = [] # initialize list for y axis containing the approximation
while(1):
 x = 4*(-1)**n/(2*n+1)
 sum += x
 n_list.append(n)
 y_list.append(sum)
 if (abs(x) < eps): break
 n += 1
```

```python
# Plot the results
plt.plot(n_list,y_list)
plt.show()
```

```python
# Plot the results along with the true values
pi_list = np.pi*np.ones(len(n_list)) # generate array with the values of all elments filled with pi
plt.plot(n_list, y_list, n_list, pi_list)
plt.show()
```

```python
# 뉴턴방법

# Another version
import numpy as np
import matplotlib.pyplot as plt
# Define the function
def f(x):
 return np.sin(x) - 0.5
# Initialize the variables
xl = -1.5
xr = 1.5
ep = 0.001
xstar = [] # List to hold the xm values
# Bisection method
while abs(xl - xr) > ep:
 xm = (xl + xr) / 2
 if f(xr) * f(xm) > 0:
  xr = xm
 else:
  xl = xm
  xstar.append(xm) # Save the xm for each iteration
# Output the final xm and f(xm)
print("Final xm:", xm)
print("f(xm):", f(xm))
# Plotting xm over iterations
plt.plot(xstar,'.')
plt.xlabel('Iteration')
plt.ylabel('xm')
plt.title('Convergence of xm')
plt.show()
```

```python
import numpy as np, matplotlib.pyplot as plt
def f(x):
 return np.sin(x)-0.5
def df(x):
 return np.cos(x)
eps = 1e-8
x = 1
steps = []
x_list = []
step = 0
while (np.abs(f(x)) > eps):
 x -= f(x)/df(x)
 steps.append(step)
 x_list.append(x)
 step += 1
plt.plot(steps, x_list)
plt.xlabel("Step")
plt.ylabel("x")
plt.yscale("log") # Display y axis in log scale
plt.show()
```

```python
# Another version
import numpy as np
def f(x):
 return np.sin(x)-0.5
def dfdx(x):
 return np.cos(x)
def NextPoint(x):
 return x-f(x)/dfdx(x)
x = 1
ep = 0.000001
xstar=[] # List to hold the xm values
while abs(f(x))>ep:
 x = NextPoint(x)
 xstar.append(x) # Save the x for each iteration
print(x)
# Plotting x over iterations
plt.plot(xstar,'.')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.title('Convergence of x')
plt.show()
```

```python
#Secant 방법

import numpy as np, matplotlib.pyplot as plt
def f(x):
 return np.sin(x)-0.5
def df(x0,x1):
 return (f(x1)-f(x0))/(x1-x0)
eps = 1e-8
x0 = 0.2
x1 = 1
steps = []
x_list = []
step = 0
x = x1 #initialize x
while (np.abs(f(x1)) > eps):
 x -= f(x)/df(x0,x1) # Calculate the next approximation of the root
 x0 = x1
 x1 = x
 steps.append(step)
 x_list.append(x)
 step += 1
# Output the final point and its function value
print("Final x:", x1)
print("f(x):", f(x1))
plt.plot(steps, x_list)
plt.xlabel("Step")
plt.ylabel("x")
plt.yscale("log") # Display y axis in log scale
plt.show()
```

```python
# Another version
import numpy as np, matplotlib.pyplot as plt
def f(x):
 return np.sin(x)-0.5
def dfdx(x0,x1):
 return (f(x1)-f(x0))/(x1-x0)
def NextPoint(x0,x1):
 return x1-f(x1)/dfdx(x0,x1)
x0 = 0
x1 = 1
ep = 1e-8
xstar=[] # List to hold the xm values
while abs(f(x1))>ep:
 x = NextPoint(x0,x1)
 x0 = x1
 x1 = x
 xstar.append(x) # Save the x for each iteration
# Output the final point and its function value
print("Final x:", x1)
print("f(x):", f(x1))
# Plotting x over iterations
plt.plot(xstar)
plt.xlabel('Iteration')
plt.ylabel('x')
plt.title('Convergence of x')
plt.show()
```

```python
# Newton's method using scipy package
from scipy.optimize import newton
import numpy as np
def f(x):
 return np.sin(x) - 0.5
def df(x):
 return np.cos(x)
# Apply Newton's method to find the root of f(x).
root = newton(f, 1.0, fprime=df)
# Print the found root to the console.
print("Root found:", root)
```
