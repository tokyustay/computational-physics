
##hist2d()
plt.hist2d() 함수는 Matplotlib 라이브러리를 사용하여 2차원 히스토그램을 생성합니다. 2차원 히스토그램은 두 변수 간의 분포를 시각화하는 데 사용되며, 각 셀(또는 빈)에 속하는 데이터 포인트의 수를 색상으로 표현합니다.

주요 파라미터:

x, y: 데이터 포인트의 x 좌표와 y 좌표를 나타내는 배열입니다.
bins: x축과 y축에 대한 빈(bin)의 수를 정의합니다. 더 많은 빈을 사용하면 더 세밀한 분포를 볼 수 있지만, 너무 많으면 데이터가 희박해질 수 있습니다.
cmap: 사용할 컬러맵을 지정합니다. 데이터의 밀도에 따라 색상이 변화합니다.
##boxplot(X)
plt.boxplot(X) 함수는 데이터의 분포를 요약하는 박스 플롯을 생성합니다. 박스 플롯은 중앙값, 사분위수(25%, 75%), 이상치 등을 시각화하여 데이터의 분포를 간략하게 요약해 줍니다.

주요 파라미터:

X: 박스 플롯을 그리기 위한 데이터입니다. 리스트의 리스트 형태로 여러 그룹의 데이터를 동시에 표시할 수 있습니다.
notch: True로 설정하면 중앙값에 대한 신뢰 구간을 표시하는 노치가 박스에 추가됩니다.
vert: True로 설정하면 박스를 세로로 그립니다. False로 설정하면 가로로 그립니다.
##pie(x)
plt.pie(x) 함수는 파이 차트를 생성하여 데이터의 상대적 비율을 원형 차트로 시각화합니다. 파이 차트는 전체에 대한 각 부분의 비율을 나타내는 데 적합합니다.

주요 파라미터:

x: 파이 차트에 표시될 데이터의 배열입니다. 각 요소는 파이의 한 조각의 크기를 결정합니다.
labels: 파이의 각 조각에 대한 레이블을 지정합니다.
autopct: 조각의 비율을 표시하는 데 사용되는 문자열 포맷입니다. 예를 들어, '%1.1f%%'는 각 조각의 비율을 백분율로 표시합니다.
startangle: 파이 차트가 그려지기 시작하는 각도입니다. 기본값은 0이며, 3시 방향에서 시작합니다.
이 함수들은 데이터의 다양한 측면을 시각화하기 위해 Matplotlib 라이브러리에서 제공하는 강력한 도구입니다. 각 함수는 데이터를 다른 방식으로 요약 및 표현하여, 데이터 분석 과정에서 유용한 인사이트를 제공할 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

wavelength = 15.0
k = np.pi*2/wavelength

side = 100
points = 500
spacing = side / points

A= np.empty([points, points], float) # points는 배열의 각 차원의 크기를 나타내며,
# float는 배열이 저장할 데이터의 타입을 지정합니다.
# 따라서, 이 구문은 points x points 크기의 2차원 실수형(float) 배열을 생성합니다.

for i in range(points):
  y = spacing*i
  for j in range(points) :
    x = spacing * j
    r = np.sqrt(x**2 + y**2)
    A[i, j] = np.sin(k*r)

plt.imshow(A, origin="lower",extent=[0,side,0,side])
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
plt.colorbar()
plt.gray()

plt.show()


#plt.jet()
#plt.hsv()
#plt.viridis()
#plt.hot()
#plt.spectral()
#plt.bone()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 파형 파라미터 정의
wavelength = 5.0  # 파장
k = 2 * np.pi / wavelength  # 파수

# 그리드 파라미터 정의
side = 100.0  # 그리드의 한 변의 길이
points = 500  # 한 변의 점 수
spacing = side / points  # 점 간격

# 두 파원의 위치
source1 = (side / 4, side / 2)
source2 = (3 * side / 4, side / 2)

# 빈 배열 초기화
A = np.zeros([points, points], float)

# 파형 패턴 생성
for i in range(points):
    y = spacing * i
    for j in range(points):
        x = spacing * j
        # 각 점에서 두 파원까지의 거리 계산
        r1 = np.sqrt((x - source1[0])**2 + (y - source1[1])**2)
        r2 = np.sqrt((x - source2[0])**2 + (y - source2[1])**2)
        # 두 구형 파의 간섭
        A[i, j] = np.sin(k * r1) + np.sin(k * r2)

# 파형 패턴 플로팅
plt.imshow(A, origin='lower', extent=[0, side, 0, side], cmap='gray')
plt.colorbar()
plt.title('Interference of Two Spherical Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 조건 및 함수 정의
eps = 1e-8
def f(x):
    return np.sin(x) - 0.5

x1 = 0
x2 = np.pi / 2.0

# x3의 값과 f(x3)의 절댓값을 저장할 리스트
x3_values = []
f_x3_values = []

# 이분법을 이용한 근 찾기
while abs(x2 - x1) > eps:
    x3 = (x1 + x2) / 2.0
    if f(x1) * f(x3) < 0:
        x2 = x3
    else:
        x1 = x3
    x3_values.append(x3)
    f_x3_values.append(abs(f(x3)))

# 근의 변화를 그래프로 표시
plt.figure(figsize=(10, 6))
plt.plot(x3_values, f_x3_values, 'o-')
plt.yscale('log')  # 세로축을 로그 스케일로 설정
plt.xlabel('Iteration')
plt.ylabel('Absolute value of f(x3)')
plt.title('Convergence of Root Finding Using Bisection Method')
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
#5
def f(x, u):
    return np.tanh(x) - u

def df(x):
    return 1 - np.tanh(x)**2

def newtons_method(u, x0=0, max_iter=1000, tol=1e-10):
    x = x0
    for _ in range(max_iter):
        x_new = x - f(x, u) / df(x)
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x

def secant_method(u, x0=0, x1=1, max_iter=1000, tol=1e-10):
    for _ in range(max_iter):
        x_new = x1 - f(x1, u) * (x1 - x0) / (f(x1, u) - f(x0, u))
        if abs(x_new - x1) < tol:
            break
        x0, x1 = x1, x_new
    return x_new

u_values = np.linspace(-0.99, 0.99, 100)

x_newtons = [newtons_method(u) for u in u_values]
x_secant = [secant_method(u) for u in u_values]

plt.figure(figsize=(12, 6))
plt.plot(u_values, x_newtons, label="Newton's Method", linestyle='-', marker='o', markersize=4)
plt.plot(u_values, x_secant, label="Secant Method", linestyle='--', marker='x', markersize=4)
plt.title("Inverse Hyperbolic Tangent (tanh⁻¹(u)) via Newton's and Secant Methods")
plt.xlabel('u')
plt.ylabel('x (tanh⁻¹(u))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6
def f(m, T):
    return np.tanh(m / T) - m

def df(m, T):
    return 1/T * (1 / np.cosh(m / T))**2 - 1

def newtons_method(T, m0=1.0, max_iter=1000, tol=1e-10):
    m = m0
    for _ in range(max_iter):
        m_new = m - f(m, T) / df(m, T)
        if abs(m_new - m) < tol:
            return m_new
        m = m_new
    return m
def bisection_method(T, m_lower=-1.0, m_upper=1.0, max_iter=1000, tol=1e-10):
    for _ in range(max_iter):
        m_mid = (m_lower + m_upper) / 2
        if f(m_mid, T) * f(m_lower, T) < 0:
            m_upper = m_mid
        else:
            m_lower = m_mid
        if abs(m_upper - m_lower) < tol:
            return (m_lower + m_upper) / 2
    return (m_lower + m_upper) / 2

temperatures = np.linspace(0.1, 1.5, 100)

magnetization_newton = np.array([newtons_method(T) for T in temperatures])
magnetization_bisection = np.array([bisection_method(T) for T in temperatures])

Tc_theoretical = 1
plt.figure(figsize=(12, 6))
plt.plot(temperatures, magnetization_newton, label="Newton's Method", marker='o')
plt.plot(temperatures, magnetization_bisection, label="Bisection Method", marker='x')
plt.axvline(x=Tc_theoretical, color='r', linestyle='--', label="Theoretical Critical Temperature (Tc)")
plt.title('Magnetization (m) as a function of Temperature (T)')
plt.xlabel('Temperature (T)')
plt.ylabel('Magnetization (m)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#homework
def f(z):
    return z**3 - 1

def df(z):
    return 3 * z**2

def newtons_method_complex(z0, max_iter=30, tol=1e-10):
    z = z0
    path = [z] # to record the path of convergence
    for _ in range(max_iter):
        z_new = z - f(z) / df(z)
        path.append(z_new)
        if abs(z_new - z) < tol:
            return z_new, path
        z = z_new
    return z, path

roots = [1, np.exp(1j * 2 * np.pi / 3), np.exp(1j * 4 * np.pi / 3)]

initial_values = [0.5 + 0.5j, -1 + 0.5j, -0.5 - 0.5j]

converged_roots = []
paths = []
for z0 in initial_values:
    root, path = newtons_method_complex(z0)
    converged_roots.append(root)
    paths.append(path)

plt.figure(figsize=(8, 8))
for i, path in enumerate(paths):
    path = np.array(path)
    plt.plot(path.real, path.imag, label=f'Start: {initial_values[i]}', marker='o')

for root in roots:
    plt.plot(root.real, root.imag, 'kx', markersize=10)

plt.title('Convergence in the Complex Plane Using Newton\'s Method')
plt.xlabel('Real part')
plt.ylabel('Imaginary part')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
